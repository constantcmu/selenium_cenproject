```python
# --- Cell 1: Imports และการตั้งค่า ---
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager


import time
import os
from dotenv import load_dotenv

# โหลดค่าจากไฟล์ .env
load_dotenv()

# --- ค่าคงที่ ---
LOGIN_URL = "https://bis.rid.go.th/report_bis.asp"
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.maximize_window()

print(f"กำลังเข้าสู่หน้า Login: {LOGIN_URL}")
driver.get(LOGIN_URL)

wait = WebDriverWait(driver, 5) # เพิ่มเวลารอเป็น 20 วินาทีเผื่อเว็บโหลดช้า
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException

def select_chosen_dropdown_option(driver, dropdown_xpath, option_text):

    try:
        # 1. รอและคลิกที่กล่อง dropdown หลักเพื่อเปิดรายการตัวเลือก
        dropdown_container = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, dropdown_xpath))
        )
        dropdown_container.click()
        
        # 2. รอและค้นหาตัวเลือก (option) ที่เป็นแท็ก <li> ซึ่งมีข้อความตรงกับที่ต้องการ
        #    ใช้ normalize-space() เพื่อจัดการกับช่องว่างที่อาจไม่เท่ากัน
        option_list_xpath = f"//li[normalize-space()='{option_text}']"
        
        option_to_select = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, option_list_xpath))
        )
        
        # 3. คลิกที่ตัวเลือกนั้น
        option_to_select.click()
        
        print(f"เลือก '{option_text}' จาก dropdown '{dropdown_xpath}' สำเร็จ")
        return True
    except (TimeoutException, NoSuchElementException) as e:
        print(f"ไม่สามารถเลือก '{option_text}' จาก dropdown '{dropdown_xpath}' ได้: {e}")
        return False

```

```python

# 1. การเลือก "ปี" จาก Dropdown 'BG_YR_chosen'
xpath_year = '//*[@id="BG_YR_chosen"]'
year_to_select = '2567' # ตัวอย่างปีที่ต้องการเลือก
select_chosen_dropdown_option(driver, xpath_year, year_to_select)

# 2. การเลือก "รหัสพื้นที่" จาก Dropdown 'GFAREA_CODE_chosen'
xpath_area_code = '//*[@id="GFAREA_CODE_chosen"]'
area_code_to_select = 'P5200 ลำปาง' # ตัวอย่างรหัสพื้นที่ที่ต้องการเลือก
select_chosen_dropdown_option(driver, xpath_area_code, area_code_to_select)
```
